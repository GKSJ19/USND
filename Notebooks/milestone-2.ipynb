```python
df['Year'] = df['Declaration Date'].dt.year
df['Month'] = df['Declaration Date'].dt.month
```

```python
import matplotlib.pyplot as plt

year_data = df.groupby('Year').size()

plt.figure(figsize=(10,5))
plt.plot(year_data.index, year_data.values)
plt.title("Total Disaster Declarations Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.show()
```

```python
disaster_year = df.groupby(['Year','Disaster Type']).size().unstack()

disaster_year.plot(figsize=(12,6))
plt.title("Disaster Types Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.legend(title="Disaster Type")
plt.show()
```

```python
month_data = df.groupby('Month').size()

plt.figure(figsize=(8,4))
plt.bar(month_data.index, month_data.values)
plt.title("Disaster Declarations by Month")
plt.xlabel("Month")
plt.ylabel("Number of Declarations")
plt.show()
```
