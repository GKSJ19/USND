```python
import pandas as pd
df = pd.read_csv('../usnd_raw.csv')
df.info()
```

```python
df = df.rename(columns={
    'Declaration Number': 'declarationNumber',
    'Declaration Type': 'declarationType',
    'Declaration Date': 'declarationDate',
    'State': 'state',
    'County': 'county',
    'Disaster Type': 'incidentType',
    'Disaster Title': 'incidentTitle',
    'Start Date': 'startDate',
    'End Date': 'endDate',
    'Close Date': 'closeDate',
    'Individual Assistance Program': 'individualAssistanceProgram',
    'Individuals & Households Program': 'individualsAndHouseholdsProgram',
    'Public Assistance Program': 'publicAssistanceProgram',
    'Hazard Mitigation Program': 'hazardMitigationProgram'
})
df.info()
```

```python
# 1. Convert declarationDate to a datetime object 
df['declarationDate'] = pd.to_datetime(df['declarationDate'])

# 2. Extract Year and Month (Required for Milestone 3) [cite: 26, 27]
df['year'] = df['declarationDate'].dt.year
df['month'] = df['declarationDate'].dt.month

# 3. Standardize State values 
df['state'] = df['state'].astype(str).str.upper().str.strip()

# 4. Check for missing values 
print("Missing values per column:\n", df.isnull().sum())

# 5. Drop rows where critical geographic info is missing 
df = df.dropna(subset=['state', 'incidentType'])
print(f"Dataset covers {df['state'].nunique()} unique states.")
```

```python
df.to_csv('usnd_cleaned.csv', index=False)
```
